```python
import pandas as pn
data=pn.read_csv("C:/Users/Moshiur/Desktop/Moshiur/Promoter/Promoter.csv",header=None)
import numpy as np
data=np.asarray(data)
data = np.delete(data, (0), axis=0)
X=data[:,:data.shape[1]-1]
y=data[:,-1:]
X
```

```python
X[:2860,[1][0]]
count=[]
for i in range(81):
    ACGT=[1.0,1.0,1.0,1.0]
    for j in range(2860):
        if X[j,1][i]=='a':ACGT[0] +=1.0
        elif X[j,1][i]=='c':ACGT[1]+=1.0
        elif X[j,1][i]=='g':ACGT[2]+=1.0
        elif X[j,1][i]=='t':ACGT[3]+=1.0
    ACGT[0] /=2864
    ACGT[1] /=2864
    ACGT[2] /=2864
    ACGT[3] /=2864
    count.append(ACGT)

feature1=[]
for i in range(2860*2):
    ar=[]
    for j in range(81):
        if X[i,1][j]=='a' or X[i,1][j]=='A':ar.append(count[j][0])
        elif X[i,1][j]=='c' or X[i,1][j]=='C':ar.append(count[j][1])
        elif X[i,1][j]=='g' or X[i,1][j]=='G':ar.append(count[j][2])
        else :ar.append(count[j][3])
    feature1.append(ar)
feature1=np.asarray(feature1)
feature1 = feature1.reshape(2860*2,81)
feature1[1,:]
```

```python
X[:2860,[1][0]]
count2=[]
for i in range(81):
    ACGT=[1.0,1.0,1.0,1.0]
    for j in range(2860):
        if X[j+2860,1][i]=='A':ACGT[0] +=1.0
        elif X[j+2860,1][i]=='C':ACGT[1]+=1.0
        elif X[j+2860,1][i]=='G':ACGT[2]+=1.0
        elif X[j+2860,1][i]=='T':ACGT[3]+=1.0
    ACGT[0] /=2864
    ACGT[1] /=2864
    ACGT[2] /=2864
    ACGT[3] /=2864
    count2.append(ACGT)

feature2=[]
for i in range(2860*2):
    ar=[]
    for j in range(81):
        if X[i,1][j]=='a' or X[i,1][j]=='A':ar.append(count2[j][0])
        elif X[i,1][j]=='c' or X[i,1][j]=='C':ar.append(count2[j][1])
        elif X[i,1][j]=='g' or X[i,1][j]=='G':ar.append(count2[j][2])
        else :ar.append(count2[j][3])
    feature2.append(ar)
feature2=np.asarray(feature2)
feature2 = feature2.reshape(2860*2,81)
feature2[1,:]
```

```python
feature=np.concatenate((feature1,feature2,y),axis=1)
feature[0,:]
dataframe = pn.DataFrame(feature)
dataframe.to_csv("BPBfeature.csv")
#pn.DataFrame(data=all_promoter[1:,1:],index=data[1:,0],columns=all_promoter[0,1:])
dataframe.head()
```
