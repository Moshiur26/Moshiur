```python
import pandas as pn
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
data=pn.read_csv("diabetes.csv",header=None)
X = pn.read_csv("X.csv",header=None)
#remove first row (header)
data = data.iloc[1:]
import numpy as np
data=np.asarray(data)
X=data[:,:data.shape[1]-1]
y=data[:,-1:]
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33,random_state=42)

test_set_size = len(y_test)
test_set_size
```

```python
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial',C=1)
clf.fit(X_train, y_train)
k=clf.predict(X_test[:, :])

columns = ['Class Value']
#df = pd.DataFrame(ex_dict, columns=columns, index=index)
df1 = pn.DataFrame(k,columns=columns)
df1
```

```python

columns = ['Real Class Value']
#df = pd.DataFrame(ex_dict, columns=columns, index=index)
df2 = pn.DataFrame(y_test,columns=columns)
df2
```

```python
clf.score(X_train, y_train)
```

```python
df = pn.concat([df1, df2], axis=1)
df
```

```python
df.to_csv("out.csv")
```
