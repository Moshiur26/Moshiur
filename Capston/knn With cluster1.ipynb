```python
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import precision_recall_fscore_support
```

```python
dataset_name="yeast.csv"#"hayes-roth.csv"
path ="E:/Subject/Capston/Farid Sir/Dataset/Classification-Datasets-master/"
data = pd.read_csv(path+dataset_name)
data=np.asarray(data)
X = data[:,:data.shape[1]-1]
#X=X.astype(np.float)
X=X.astype(np.int)
y = data[:,-1:]
y = np.unique(y, return_inverse=True)[1].tolist()
number_of_lavel=len(np.unique(y))
number_of_cluster=len(np.unique(y))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=190)
kmeans = KMeans(n_clusters=number_of_cluster, random_state=0).fit(X_train)
len(X_train)
X_train[0]
number_of_lavel

#y
```

```python
#kmeans.labels_
```

```python
cluster=np.zeros((number_of_cluster,number_of_lavel))
#cluster
```

```python
#cluster_predict=kmeans.predict(X_train)
cluster_predict=kmeans.labels_
#cluster_predict
```

```python
for i in range(len(X_train)):
    cluster[cluster_predict[i]][y[i]-1] +=1
cluster[0]
```

```python
label=np.zeros(number_of_cluster)
cluster[1].max()
label
```

```python
for i in range(len(label)):
    for j in range(number_of_lavel):
        if cluster[i].max()==cluster[i][j]:
            label[i]=j
            break
label
```

```python
success=0.0
unsuccess=0.0
predict_test=kmeans.predict(X_test)
for i in range(len(y_test)):
    if label[predict_test[i]]==y_test[i]: 
        success+=1
    else:
        unsuccess+=1
    print(predict_test[i],"==",y_test[i])
success/(success+unsuccess)
```
