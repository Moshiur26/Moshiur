```python
import numpy as np
import pandas as pn
```

```python
data=pn.read_csv("E:/AI/Py Project/disease/Gaji (Disease)/Training.csv",header=None)
data=np.asarray(data)
feature_name=data[0,:]
data = np.delete(data, (0), axis=0)
X=data[:,:data.shape[1]-1]
y=data[:,-1:]
X=X.astype(int)
X_len=len(X)

#s=X.sum(axis=1).sum(axis=0)
#X.sum(axis=1)
X[0]

#feature_name[len(X[0])-1]
```

```python
train_dataset=[]
for i in range(X_len):
    arr=[]
    arr.append(y[i][0])
    for j in range(len(X[0])):
        if X[i][j]==1:
            arr.append(feature_name[j])
    train_dataset.append(arr)
train_dataset[X_len-1]
```

```python
dataframe = pn.DataFrame(train_dataset)
dataframe.to_csv("Train.csv")
#pn.DataFrame(data=all_promoter[1:,1:],index=data[1:,0],columns=all_promoter[0,1:])
dataframe.head()
```
